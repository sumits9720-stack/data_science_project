```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Part A — Data preparation
```

```python
fear_greed_df = pd.read_csv('fear_greed_index.csv')
historical_data_df = pd.read_csv('historical_data.csv')

print("Rows and Columns")
print(f"Fear & Greed Index shape: {fear_greed_df.shape}")
print(f"Historical Data shape: {historical_data_df.shape}\n")

print("Missing Values\n")
print("Fear & Greed Index:\n", fear_greed_df.isnull().sum())
print("\nHistorical Data:\n", historical_data_df.isnull().sum())

print("Duplicates")
print(f"Fear & Greed Index duplicates: {fear_greed_df.duplicated().sum()}")
print(f"Historical Data duplicates: {historical_data_df.duplicated().sum()}")
```

```python
fear_greed_df['date'] = pd.to_datetime(fear_greed_df['date'])

historical_data_df['Timestamp IST'] = pd.to_datetime(historical_data_df['Timestamp IST'], dayfirst=True)
historical_data_df['date'] = pd.to_datetime(historical_data_df['Timestamp IST'].dt.date)

merged_df = pd.merge(
    historical_data_df, 
    fear_greed_df[['date', 'value', 'classification']], 
    on='date', 
    how='left'
)
merged_df.rename(columns={'value': 'fear_greed_value'}, inplace=True)
```

```python
print(merged_df.head(2))
```

```python
daily_pnl_per_trader = merged_df.groupby(['date', 'Account'])['Closed PnL'].sum().reset_index()

closed_trades = merged_df[merged_df['Closed PnL'] != 0]
win_rate = (closed_trades['Closed PnL'] > 0).mean() * 100
avg_trade_size = merged_df['Size USD'].mean()

print(f"\nOverall Win Rate: {win_rate:.2f}%")
print(f"Overall Average Trade Size: ${avg_trade_size:.2f}")

trades_per_day = merged_df.groupby('date').size().reset_index(name='Daily Trade Count')

buys_per_day = merged_df[merged_df['Side'].str.upper() == 'BUY'].groupby('date').size()
sells_per_day = merged_df[merged_df['Side'].str.upper() == 'SELL'].groupby('date').size()

long_short_df = pd.DataFrame({'Buys': buys_per_day, 'Sells': sells_per_day}).fillna(0)
long_short_df['Long_Short_Ratio'] = long_short_df['Buys'] / long_short_df['Sells'].replace(0, 1)
long_short_df = long_short_df.reset_index()

print("\n Daily Long/Short Ratio Preview")
print(long_short_df.head())
```

```python
daily_pnl_per_trader.to_csv('daily_pnl_per_trader.csv', index=False)
trades_per_day.to_csv('trades_per_day.csv', index=False)
long_short_df.to_csv('daily_long_short_ratio.csv', index=False)
merged_df.to_csv('merged_trading_sentiment.csv', index=False)
```

```python
#Part B — Analysis
```

```python
merged_df = pd.read_csv('merged_trading_sentiment.csv')
merged_df['date'] = pd.to_datetime(merged_df['date'])


def categorize_sentiment(classification):
    if pd.isna(classification):
        return 'Unknown'
    classification = classification.lower()
    if 'fear' in classification:
        return 'Fear'
    elif 'greed' in classification:
        return 'Greed'
    else:
        return 'Neutral'

merged_df['Sentiment_Category'] = merged_df['classification'].apply(categorize_sentiment)

# Filter out zero PnL trades for win rate calculation
closed_trades = merged_df[merged_df['Closed PnL'] != 0].copy()
closed_trades['Is_Win'] = closed_trades['Closed PnL'] > 0

# Calculate metrics by sentiment
perf_by_sentiment = closed_trades.groupby('Sentiment_Category').agg(
    Total_PnL=('Closed PnL', 'sum'),
    Average_PnL=('Closed PnL', 'mean'),
    Win_Rate=('Is_Win', 'mean'),
    Trade_Count=('Closed PnL', 'count')
).reset_index()
perf_by_sentiment['Win_Rate'] = perf_by_sentiment['Win_Rate'] * 100

print("--- Performance by Sentiment ---")
print(perf_by_sentiment)

behavior_by_sentiment = merged_df.groupby('Sentiment_Category').agg(
    Avg_Trade_Size=('Size USD', 'mean'),
    Total_Trades=('Size USD', 'count')
).reset_index()

# Long/Short ratio by sentiment
buys_by_sentiment = merged_df[merged_df['Side'].str.upper() == 'BUY'].groupby('Sentiment_Category').size()
sells_by_sentiment = merged_df[merged_df['Side'].str.upper() == 'SELL'].groupby('Sentiment_Category').size()
ls_ratio = (buys_by_sentiment / sells_by_sentiment.replace(0, 1)).reset_index(name='Long_Short_Ratio')

behavior_by_sentiment = pd.merge(behavior_by_sentiment, ls_ratio, on='Sentiment_Category', how='left')

print("\n--- Behavior by Sentiment ---")
print(behavior_by_sentiment)

trader_stats = closed_trades.groupby('Account').agg(
    Total_Trades=('Closed PnL', 'count'),
    Total_PnL=('Closed PnL', 'sum'),
    Win_Rate=('Is_Win', 'mean'),
    Avg_Trade_Size=('Size USD', 'mean')
).reset_index()

freq_threshold = trader_stats['Total_Trades'].quantile(0.75)
trader_stats['Frequency_Segment'] = np.where(trader_stats['Total_Trades'] >= freq_threshold, 'Frequent', 'Infrequent')

trader_stats['Consistency_Segment'] = np.where((trader_stats['Win_Rate'] >= 0.6) & (trader_stats['Total_PnL'] > 0), 'Consistent Winner', 'Inconsistent/Loser')

print("\n--- Trader Segments Breakdown (Frequency) ---")
print(trader_stats['Frequency_Segment'].value_counts())

print("\n--- Trader Segments Breakdown (Consistency) ---")
print(trader_stats['Consistency_Segment'].value_counts())
```

```python
#Total PnL by Sentiment
plt.figure(figsize=(8, 4))
sns.barplot(data=perf_by_sentiment, x='Sentiment_Category', y='Total_PnL', hue='Sentiment_Category', legend=False, palette='viridis')
plt.title('Total PnL by Market Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Total PnL (USD)')
plt.show() # <-- Changed to show()

# Win Rate by Sentiment
plt.figure(figsize=(8, 4))
sns.barplot(data=perf_by_sentiment, x='Sentiment_Category', y='Win_Rate', hue='Sentiment_Category', legend=False, palette='coolwarm')
plt.title('Win Rate by Market Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Win Rate (%)')
plt.ylim(0, 100)
plt.show() # <-- Changed to show()

# Average Trade Size by Sentiment
plt.figure(figsize=(8, 4))
sns.barplot(data=behavior_by_sentiment, x='Sentiment_Category', y='Avg_Trade_Size', hue='Sentiment_Category', legend=False, palette='magma')
plt.title('Average Trade Size by Market Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Average Trade Size (USD)')
plt.show() # <-- Changed to show()

#Average PnL by Trader Consistency
seg_pnl = trader_stats.groupby('Consistency_Segment')['Total_PnL'].mean().reset_index()
plt.figure(figsize=(8, 4))
sns.barplot(data=seg_pnl, x='Consistency_Segment', y='Total_PnL', hue='Consistency_Segment', legend=False, palette='Set2')
plt.title('Average PnL by Trader Consistency')
plt.xlabel('Segment')
plt.ylabel('Average Total PnL (USD)')
plt.show() # <-- Changed to show()
```
